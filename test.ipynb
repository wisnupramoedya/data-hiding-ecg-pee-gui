```python
text = "HaloWorld"

# Encode the text as bytes using UTF-8 encoding
encoded_bytes = text.encode('utf-8')

# Convert each byte to a binary string representation
binary_strings = [format(byte, '08b') for byte in encoded_bytes]

# Join the binary strings to get a single string of bits
bit_string = ''.join(binary_strings)

print("String of Bits:", bit_string)
```

```python
bit_string = "01010011011000010111100101100001001000000110100101101110011001110110100101101110001000000110110101100001011010110110000101101110"

# Split the bit string into 8-bit chunks
bit_chunks = [bit_string[i:i+8] for i in range(0, len(bit_string), 8)]

# Convert each 8-bit chunk to an integer
bytes_list = [int(chunk, 2) for chunk in bit_chunks]

# Convert the list of integers to bytes
byte_array = bytearray(bytes_list)

# Decode the bytes using UTF-8 encoding to get the original string
original_string = byte_array.decode('utf-8')

print("Decoded String:", original_string)
```

```python
import numpy as np

start = -6.000
stop = 6.000
step = 0.001

# Membuat array NumPy
arr = np.arange(start, stop + step, step)

# Cetak array yang telah dibuat
print(arr[:10])
print(arr.dtype)
```

```python
print(f'{arr[2]}')
```

```python
def preproccess_signal(original_data: np.ndarray) -> np.ndarray:
    return ((np.round(original_data, 3) * 1000).flatten()).astype(np.int64)


def postproccess_signal(original_data: np.ndarray) -> np.ndarray:
    # Divide each element by 1000 and reshape to a 2D array
    output_array = original_data / 1000.0  # Divide by 1000 to get decimal values
    output_array = output_array.reshape(-1, 1)  # Reshape to a single column

    return output_array
```

```python
arr_pre = preproccess_signal(arr)
print(arr_pre)
arr2 = postproccess_signal(arr_pre).flatten()
print(arr2)
```

```python
print(arr2[2])
```

```python
def compare_arrays(arr1, arr2):
    # Memeriksa apakah bentuk (shape) kedua array sama
    if arr1.shape != arr2.shape:
        return "Bentuk (shape) kedua array berbeda"

    # Mencari indeks dan nilai yang berbeda
    indices, = np.where(arr1 != arr2)
    values = arr1[indices]
    values2 = arr2[indices]

    # Mengembalikan hasil berupa pasangan indeks dan nilai yang berbeda
    return list(zip(indices, values, values2))


hasil_perbandingan = compare_arrays(np.round(arr, 3), arr2)

if isinstance(hasil_perbandingan, list):
    if len(hasil_perbandingan) > 0:
        print("Indeks yang berbeda dan nilainya:")
        for indeks, nilai, nilai2 in hasil_perbandingan:
            print(f"Indeks: {indeks}, Nilai: {nilai} Nilai2: {nilai2}")
    else:
        print("Tidak ada perbedaan antara kedua array.")
else:
    print(hasil_perbandingan)
```

```python
arr_pre2 = preproccess_signal(arr2)
print(arr_pre2)
arr3 = postproccess_signal(arr_pre2).flatten()
print(arr3)

hasil_perbandingan = compare_arrays(arr2, arr3)

if isinstance(hasil_perbandingan, list):
    if len(hasil_perbandingan) > 0:
        print("Indeks yang berbeda dan nilainya:")
        for indeks, nilai, nilai2 in hasil_perbandingan:
            print(f"Indeks: {indeks}, Nilai: {nilai} Nilai2: {nilai2}")
    else:
        print("Tidak ada perbedaan antara kedua array.")
else:
    print(hasil_perbandingan)
```
